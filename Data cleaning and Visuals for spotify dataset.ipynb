```python
target_columns = [
    'YouTube Views', 'TikTok Posts', 'YouTube Likes', 'TikTok Likes', 'TikTok Views', 
    'YouTube Playlist Reach', 'Apple Music Playlist Count', 'AirPlay Spins', 'SiriusXM Spins', 
    'Deezer Playlist Count', 'Deezer Playlist Reach', 'Amazon Playlist Count', 'Pandora Streams', 
    'Pandora Track Stations', 'Soundcloud Streams', 'Shazam Counts', 'TIDAL Popularity'
]

imputer = SimpleImputer(strategy='most_frequent')


for column in target_columns:
    if df[column].isnull().any():  
        if df[column].notnull().sum() > 0: #checking for NaN values in the columns
            df[column] = imputer.fit_transform(df[[column]].values.reshape(-1, 1)).ravel() # Re-shaping for the proper imputer
        else:
            print(f"Column {column} has all NaN values and cannot be imputed with 'most_frequent' strategy.") # Tells me which columns have all NaN values

df.isnull().sum()

#TIDAL has all NaN values so I needed to figure out a way to impute all the other missing values while figuring out which
#column had all NaN values. Honestly, I should have checked the data for NaN values before imputing missing values and leave out TIDAL 
#for a simpler string of code
```

```python
target_column = 'TIDAL Popularity'

constant_imputer = SimpleImputer(strategy='constant', fill_value=0)

df[target_column] = constant_imputer.fit_transform(df[[target_column]].values.reshape(-1, 1)).ravel()

df.isnull().sum()

#Since this column is all NaN values, you can't use the 'most_frequent' strategy for imputation because there are no values to determine the most frequent one
#Instead I used a constant value to impute these columns
```

```python

# Step 1: Convert 'Release Date' to datetime, handling errors
cleaned_df['Release Date'] = pd.to_datetime(cleaned_df['Release Date'], errors='coerce')

# Step 2: Check for rows that couldn't be converted and drop them
invalid_dates = cleaned_df['Release Date'].isna()
if invalid_dates.any():
    print("Rows with invalid 'Release Date':")
    print(cleaned_df[invalid_dates])
    # Drop rows with invalid 'Release Date'
    cleaned_df = cleaned_df[~invalid_dates]

# Step 3: Filter the data to include only tracks released from 1999 to 2024
df_filtered = cleaned_df[(cleaned_df['Release Date'].dt.year >= 1999) & (cleaned_df['Release Date'].dt.year <= 2024)].copy()

# Step 4: Extract the year from the 'Release Date'
df_filtered['Year'] = df_filtered['Release Date'].dt.year

# Step 5: Count the number of tracks released each year
tracks_per_year = df_filtered.groupby('Year').size()

# Step 6: Plot the data
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
tracks_per_year.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of Tracks Released Each Year (1999-2024)')
plt.xlabel('Year')
plt.ylabel('Number of Tracks')
plt.xticks(rotation=45)
plt.show()
```

```python
import re

def clean_text(text):
    # Remove non-printable characters and specific problematic symbols
    cleaned_text = re.sub(r'[^\x00-\x7F]', '', text)  # Remove non-ASCII characters
    cleaned_text = re.sub(r'[$]', '', cleaned_text)  # Remove specific problematic symbols
    return cleaned_text

# Apply cleaning function to your DataFrame column
df['Track'] = df['Track'].apply(clean_text)
```

```python

plt.figure(figsize=(8, 6))
sns.scatterplot(x='TikTok Posts', y='All Time Rank', data=cleaned_df, s=80, color='blue', alpha=0.7)
sns.regplot(x='TikTok Posts', y='All Time Rank', data=cleaned_df, scatter=False, color='orange', label='Trend Line')

plt.title('Relationship between TikTok Posts and Track Rank')
plt.xlabel('TikTok Posts')
plt.ylabel('Track Rank')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
